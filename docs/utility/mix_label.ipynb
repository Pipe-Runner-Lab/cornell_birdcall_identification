```python
import pandas as pd
import random
```

```python
old_csv = pd.read_csv("../../__data__/bird_song/clean_train.csv", index_col=False)
```

```python
old_csv.head()
```

```python
old_csv["mix"] = ""

for ebird_code in old_csv["ebird_code"].unique():
    inclusive_idx = old_csv.index[old_csv['ebird_code'] == ebird_code].tolist()
    exclusive_idx = old_csv.index[old_csv['ebird_code'] != ebird_code].tolist()
    
    for idx in inclusive_idx:
        old_csv.at[idx, "mix"] = random.choices(exclusive_idx, k=4)
```

```python
old_csv
```

```python
old_csv.to_csv("../../__data__/bird_song/mix_train.csv")
```
